```python
import os
import cv2
import time
import random
import numpy as np
import subprocess

import matplotlib.pyplot as plt
```

```python
def tap(x, y, tx=None, ty=None, error=15, duration: int = 0):
    x += random.uniform(-error, error)
    y += random.uniform(-error, error)
    duration += int(random.uniform(50, 100))
    if tx is None:
        tx = x
    tx += random.uniform(-error, error)
    if ty is None:
        ty = y
    ty += random.uniform(-error, error)
    os.system(f"adb shell input swipe {x} {y} {tx} {ty} {duration}")


def wait(t=0.2):
    error = random.uniform(0, 0.3)
    time.sleep(t + error)


def catch_screen():
    pipe = subprocess.Popen("adb shell screencap -p",
                            stdin=subprocess.PIPE,
                            stdout=subprocess.PIPE,
                            shell=True)
    image_bytes = pipe.stdout.read().replace(b'\r\n', b'\n')
    image = cv2.imdecode(np.frombuffer(image_bytes, np.uint8),
                         cv2.IMREAD_COLOR)
    return image

def use_skills():
    tap_skill(0)
    wait()
    tap_skill(1)
    wait()
    tap_skill(0)


def next_compaign():
    y_range = (420, 500)
    y = random.uniform(*y_range)

    x_range = (500, 600)
    x = random.uniform(*x_range)
    tap(x, y)
    wait(0)
    x_range = (700, 800)
    x = random.uniform(*x_range)
    tap(x, y)
    wait()
    tap(*pos["choose"])
    wait()
    tap(*pos["choose"])


def start_instance():
    tap(*pos["choose"])
    wait(0.8)
    tap(*pos["team"])
    wait(0.6)
    tap(*pos["choose"])
    wait(1)


def quit_instance():
    x_range = (800, 1000)
    y_range = (1040, 1080)

    x = random.uniform(*x_range)
    y = random.uniform(*y_range)
    tap(x, y)

    wait()

    x_range = (1240, 1400)
    y_range = (830, 870)

    x = random.uniform(*x_range)
    y = random.uniform(*y_range)
    tap(x, y)

    wait()

    wait(1.3, 1.5)
    tap_skill(3)
    tap_skill(3)
    wait(2, 2.2)


def battle(skills):
    for i in skills:
        tap(*pos[f"skill_{i}"])
        wait()
    tap(*pos["next_turn"])


def get_bonus():
    #奖励
    tap(*pos["skill_0"])
    wait()
    tap(*pos["bonus_confirm"])
    wait(1, 1.2)


def is_victory(threshold=0.9):
    img = catch_screen()
    tpl = cv2.imread("victory.png")

    result = cv2.matchTemplate(img, tpl, cv2.TM_CCOEFF_NORMED)
    return result.max() >= threshold
```
